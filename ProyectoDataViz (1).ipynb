### 5.9 Vamos a limpiar otro dataset (tabla) que se va a conectar a nuestro modelo, la de **Locations**
Esta tabla traduce los lugares en dónde se tomó un taxi a través de un id único la tabla de TaxiTrip

Vamos a utilizar la librería geopy que nos permite obtener las ubicaciones de los sitios específicos siempre y cuando se le ingreso el nombre de manera correcta, con esta función vamos a poder obtener las latitudes y longitudes de dónde empezó y terminó cada viaje

```python
from geopy.geocoders import Nominatim

def get_latitude_longitude(location):

    geolocator = Nominatim(user_agent="saicuuuu_bro")
    try:

        loc = geolocator.geocode(location+", New York")
        if loc:
            return (loc.latitude, loc.longitude)
        else:
            return None
    except Exception as e:
        return str(e)

print(get_latitude_longitude("Newark Airport"))
```

### 5.10 Vamos a cargar los datos de las locaciones de la tabla que nos provee: 

https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page

```python
locations_df = pd.read_csv("taxi_zone_lookup.csv")
locations_df.head()
```

```python
float_values = locations_df['Zone'].apply(lambda x: isinstance(x,float))
float_values
```

```python
def process_zones(row):
    
    cleaned_zones = []
    if '/' in row['Zone']:
        zones = row['Zone'].split('/')
        
        existing_zones = set(locations_df['cleaned_zone'].dropna())
    
        if len(zones) > 1:
            zone = zones[1].strip()  
            cleaned_zones.append(zone)
        
    else:
        zone = row['Zone'].strip()
        cleaned_zones.append(zone)
    return cleaned_zones[0] if cleaned_zones else pd.NA

locations_df['cleaned_zone'] = pd.NA

locations_df['cleaned_zone'] = locations_df.apply(process_zones, axis=1)


```

```python
locations_df
```

```python
locations_df['coordinates'] = locations_df['cleaned_zone'].apply(get_latitude_longitude)
locations_df
```

```python
# Vamos a ver si tenemos algún valor duplicado dentro de nuestras coordenadas
duplicated_values = locations_df[locations_df['coordinates'].duplicated(keep=False)]
duplicated_values
```

```python
print(locations_df[locations_df['coordinates'].isnull()])
```

```python
# Lo que se puede ver es el hecho de que hay varias zonas que no son detectadas por el hecho de que tienen
# la palabra south, north, east, west, vamos a eliminarlas porque haciendo una búsqueda previa, esto no afecta
# en gran medida el valor de las coordenadas
# En este sentido, si vamos a poder tener duplicados
words_to_remove = {'North', 'South', 'East', 'West'}
def no_nsew(row):
    if pd.isna(row['coordinates']):

        for word in words_to_remove:
            if word in row['cleaned_zone']:
                row['cleaned_zone'] = row['cleaned_zone'].replace(word, '').strip()
    return row['cleaned_zone']

locations_df['cleaned_zone'] = locations_df.apply(no_nsew, axis=1)

print(locations_df[locations_df['coordinates'].isnull()])
```

```python
null_coordinates = locations_df[locations_df['coordinates'].isnull()]

locations_df.loc[locations_df['coordinates'].isnull(), 'coordinates'] = null_coordinates['cleaned_zone'].apply(get_latitude_longitude)
```

Se puede ver que el gráfico es difícil de entender así que si quieremos evitar que en las visualizaciones de Power BI se vea igual vamos a discretir las horas


```python


def discretize_hour(time):

    minute = time.minute
    
    if minute < 8:
        return time.replace(minute=0, second=0)

    elif minute < 23:
        return time.replace(minute=15, second=0)

    elif minute < 38:
        return time.replace(minute=30, second=0)

    elif minute < 53:
        return time.replace(minute=45, second=0)

    else:
        return time + pd.Timedelta(hours=1, minutes=-minute)

taxis['discrete_time'] = taxis['tpep_pickup_datetime'].apply(discretize_hour)

    
```

```python
# Function to check if the coordinate tuple is valid
def is_valid_coordinate(coord):
    return (isinstance(coord, tuple) and len(coord) == 2 and 
            all(isinstance(x, (int, float)) for x in coord))

# Check for malformed coordinates
clean_locations['valid_coordinates'] = clean_locations['coordinates'].apply(is_valid_coordinate)
malformed_coords = clean_locations[~clean_locations['valid_coordinates']]
print("Malformed coordinates:\n", malformed_coords)
```

```python

def parse_coordinates(coord_str):
    if isinstance(coord_str, str):
        try:
            
            lat_str, lon_str = coord_str.strip("()").split(",")
            
            return float(lat_str), float(lon_str)
        except Exception as e:
            print(f"Error parsing coordinates '{coord_str}': {e}")
            return None, None
    return None, None


clean_locations[['latitude', 'longitude']] = clean_locations['coordinates'].apply(lambda x: pd.Series(parse_coordinates(x)))
```
